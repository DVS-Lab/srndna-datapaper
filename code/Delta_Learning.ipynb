```python
import pandas as pd
import os
import re
import scipy
```

```python
data_dir='../../srndna-data/stimuli/psychopy/logs'
subs=[p for p in os.listdir(data_dir) if p.startswith(('1','2'))]

f_list=[]
for sub in subs:
    f_list= f_list+[data_dir+"/%s/"%(sub)+f for f in os.listdir(data_dir+"/%s"%(sub)) if 'ultimatum' in f]

df_list=[]
for f in f_list:
    sub=re.search('logs/(.*)/',f).group(1)
    tmp=pd.read_csv(f)
    if len(tmp[tmp['Partner']=='Partner'])>0:
        x=tmp[tmp['Partner']=='Partner'].index[-1]
        tmp=pd.read_csv(f,skiprows=x)

    tmp['sub']=sub
    df_list.append(tmp)
df=pd.concat(df_list)
df.reset_index(inplace=True)
df=df[['sub','Partner','Offer','resp']]
df=df.rename(columns={'sub':"subjID",
                     "Offer":"offer",
                     "resp":"accept"})
df['accept']=pd.to_numeric(df['accept'])-2
df=df[df['accept']<=900]
df.dropna(inplace=True)
display(df.head())
for column in df.columns:
    print ("Unique values for  "+column+":",df[column].unique())
```

```python
df1=df[df['Partner']==1].copy()
df2=df[df['Partner']==2].copy()
df3=df[df['Partner']==3].copy()
df4=[]
display(df1)
for tmp in [df1,df2,df3]:
    display()
    tmp['subjID']=tmp['subjID']+'_'+tmp['Partner'].astype(int).astype(str)
    df4.append(tmp)
    
df4=pd.concat(df4)
df4['subjID'].unique()
```

```python
#from hbayesdm.diagnostics import rhat, print_fit, hdi, plot_hdi, extract_ic
from hbayesdm.models import ug_delta, ug_bayes

output1=ug_delta(data=df1,niter=2000,nwarmup=1000,nchain=10,ncore=6)
```

```python
output2=ug_delta(data=df2,niter=2000,nwarmup=1000,nchain=10,ncore=6)
```

```python
output3=ug_delta(data=df3,niter=2000,nwarmup=1000,nchain=10,ncore=6)
```
